### 3) Exploratory Data Analysis (EDA)

```python
data.drop_duplicates(inplace=True)
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Month'] = data['Order Date'].dt.month
data['Sales'] = data['Quantity Ordered'] * data['Price Each']
```

```python
data.describe()
```

```python
print(data.dtypes)
```

### 4) Data Visualizations

```python
plt.hist(data['Sales'], bins=35)
plt.xlabel('Sales')
plt.ylabel('Count')
plt.show()
```

**Monthly Sales Trend**

```python
monthly_sales = data.groupby('Month')['Sales'].sum()
months = range(1,13)

plt.figure(figsize=(15,5))
plt.bar(months, monthly_sales)
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.title('Monthly Sales Trend')
plt.xticks(months)
plt.show()
```

**Top Best Selling Products**

```python
top_products = data.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False)

plt.figure(figsize=(15,5))
top_products.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.title('Top 10 Best-Selling Products')
plt.xticks(rotation=45)
plt.show()
```

**City-Wise Sales**

```python
city_sales = data.groupby('City')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(15,5))
city_sales.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Total Sales ($)')
plt.title('City-Wise Sales')
plt.xticks(rotation=90)
plt.show()
```

**Hourly Sales Distribution**

```python
hourly_sales = data.groupby('Hour')['Sales'].sum()

plt.figure(figsize=(15,5))
plt.plot(hourly_sales.index, hourly_sales.values, marker='o')
plt.xlabel('Hour')
plt.ylabel('Total Sales ($)')
plt.title('Hourly Sales Distribution')
plt.xticks(hourly_sales.index)
plt.grid(True)
plt.show()
```
